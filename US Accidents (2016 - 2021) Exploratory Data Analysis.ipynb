Start Time

```python
df.Start_Time
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce')
```

```python
sns.distplot(df.Start_Time.dt.hour, bins=24, kde=False,norm_hist=True)
```

- A high percentage of accidents occur between 2 pm to 5 pm because of Daylight Saving Time
- Next highest percentage is 7 am to 9 am.

Which days of week have the most accidents?

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins=7, kde=False,norm_hist=True)
```

IS the distribution of accidents by hour the same on weekends as on weekdays.

```python
mondays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.distplot(mondays_start_time.dt.hour, bins=24, kde=False,norm_hist=True)
```

 What time of the day are accidents most frequent in?

```python
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(sundays_start_time.dt.hour, bins=24, kde=False,norm_hist=True)
```

On Sundays, the peak occurs between 11 am to 6 pm.

 Which months have the most accidents?

```python
sns.distplot(df.Start_Time.dt.month, bins = 12, kde=False, norm_hist=True)
```

What is the trend of accidents year over year(decreasing/increasing)?

```python
df_2021 = df[df.Start_Time.dt.year == 2021]
sns.distplot(df_2021.Start_Time.dt.month, bins=12, kde=False,norm_hist=True)
```

- much data is missing for 2020. Maybe even 2016

```python
sample_df = df.sample(int(0.001 * len(df)))
lat_lon_pairs = list(zip(list(sample_df.Start_Lat),list(sample_df.Start_Lng)))
```

```python
map = folium.Map()
for lat in lat_lon_pairs[:500]:
    marker=folium.Marker(lat,popup='GPS Location of Accidents' ,tooltip='Click here to see Popup',
                         tiles="OpenStreetMap")
    marker.add_to(map)

map
```
